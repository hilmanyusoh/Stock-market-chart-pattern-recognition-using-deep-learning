```python
from cassandra.cluster import Cluster
```

```python
# 1. เชื่อมต่อ Database
cluster = Cluster(['127.0.0.1'])
session = cluster.connect()
```

```python
# 2. สร้าง Keyspace
session.execute("""
    CREATE KEYSPACE IF NOT EXISTS stock_data 
    WITH replication = {'class': 'SimpleStrategy', 'replication_factor': '1'}
""")
session.set_keyspace('stock_data')
```

```python
# 3. สร้าง Table (เน้นใช้ date ASC เพื่อให้ข้อมูลเรียงลำดับเวลา)
session.execute("""
    CREATE TABLE IF NOT EXISTS candlestick_data (
        symbol text,
        time timestamp,
        open_price float,
        high_price float,
        low_price float,
        close_price float,
        volume bigint,
        value float,
        PRIMARY KEY (symbol, time)
    ) WITH CLUSTERING ORDER BY (time ASC);
""")
print("✅ Keyspace และ Table พร้อมใช้งาน!")
```

```python
# ลบตาราง candlestick_data ออกจาก Keyspace ปัจจุบัน (stock_data)
session.execute("DROP TABLE IF EXISTS candlestick_data")
# DROP TABLE IF EXISTS stock_data.candlestick_data;
```
